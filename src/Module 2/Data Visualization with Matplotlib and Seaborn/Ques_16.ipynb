```python
# Combining Plots with Matplotlib
# Objective: Combine multiple plots into a single figure using Matplotlib's subplots.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
scatter_y = np.cos(x)

# Create a figure with 1 row and 3 columns of subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 4))

# Line plot
axs[0].plot(x, y, color='blue')
axs[0].set_title('Line Plot')
axs[0].set_xlabel('X')
axs[0].set_ylabel('sin(X)')

# Bar plot
axs[1].bar(categories, values, color='orange')
axs[1].set_title('Bar Plot')
axs[1].set_ylabel('Values')

# Scatter plot
axs[2].scatter(x, scatter_y, color='green')
axs[2].set_title('Scatter Plot')
axs[2].set_xlabel('X')
axs[2].set_ylabel('cos(X)')

# Layout adjustment
plt.tight_layout()

# Show the combined figure
plt.show()
```
