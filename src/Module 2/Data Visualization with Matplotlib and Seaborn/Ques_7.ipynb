```python
# Advanced Scatter Plot with Seaborn
# Description: Using Seaborn, enhance the scatter plot for height vs. weight by adding size for points based
# on data: Size [10, 15, 13, 18, 12].
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Height': [160, 165, 170, 175, 180],
    'Weight': [55, 60, 65, 70, 75],
    'Size': [10, 15, 13, 18, 12]  # Custom sizes
}

# Create DataFrame
df = pd.DataFrame(data)

# Create scatter plot
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    data=df,
    x='Height',
    y='Weight',
    size='Size',              # Variable point size
    sizes=(50, 300),          # Min and max size for scaling
    legend='brief',
    palette='viridis'         # Optional: Adds a color palette for better appearance
)

# Improve plot readability
plt.title('Advanced Scatter Plot: Height vs. Weight (with Size)')
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.legend(title='Size')
plt.tight_layout()
plt.show()
```
