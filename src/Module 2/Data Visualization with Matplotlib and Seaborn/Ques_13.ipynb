```python
# Customizing Plots with Seaborn - Boxplot
# Objective: Customize a boxplot with Seaborn to show the distribution of petal lengths by
# species.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
iris = sns.load_dataset("iris")

# Set Seaborn style
sns.set(style="whitegrid")

# Create customized boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=iris,
    x='species',
    y='petal_length',
    palette='pastel',
    width=0.6,
    linewidth=2
)

# Add title and labels
plt.title('Distribution of Petal Length by Species', fontsize=14)
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.tight_layout()
plt.show()
```
