```python
# Scatter Plot of Variable Relationships
# Description: Use Seaborn to create a scatter plot showing the relationship between two
# variables.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
iris = sns.load_dataset("iris")

# Set Seaborn style
sns.set(style="whitegrid")

# Create scatter plot
plt.figure(figsize=(7, 5))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species', palette='deep')

# Add titles and labels
plt.title('Sepal Length vs. Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.tight_layout()
plt.show()
```
