```python
# Styling with Matplotlib - Line Plot
# Objective: Create a line plot with customized styles using Matplotlib.
```

```python
import matplotlib.pyplot as plt

# Sample data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [100, 120, 150, 170, 160, 180]

# Create line plot
plt.figure(figsize=(8, 5))
plt.plot(
    months,
    sales,
    marker='o',            # Add markers to points
    linestyle='--',        # Dashed line
    color='teal',          # Line color
    linewidth=2,           # Line thickness
    markersize=8           # Marker size
)

# Customize axes and title
plt.title('Monthly Sales Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales', fontsize=12)

# Add grid and style
plt.grid(True, linestyle=':', color='gray', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```
