```python
import numpy as np


data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
maximum = np.max(data)
minimum = np.min(data)


print(f"Data: {data}")
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
```
